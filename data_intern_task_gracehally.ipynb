```python
text1 = []
with open('2021_11_30.txt') as f:
  text1 = f.readlines()

text2 = []
with open('2021_12_04.txt') as f:
  text2 = f.readlines()

text3 = []
with open('2021_12_05.txt') as f:
  text3 = f.readlines()
```

```python
data = text1 + text2 + text3

data
```

```python
data= [i.split('\n')[0] for i in data]

data
```

```python
for i in data:
  if i == '----------NEW TRANSACTON-----------' or i == ' ':
    data.remove(i)
```

```python
data
```

```python
list1 =[]
list2 =[]
list3 =[]
list4 =[]
list5 =[]

for i in data:
  if 'transaction date' in i.lower():
    j = i.replace('TRANSACTION DATE : ', '')
    list1.append(j)
  elif 'client id' in i.lower():
    j = i.replace('CLIENT ID : ', '')
    list2.append(j)
  elif 'client name' in i.lower():
    j = i.replace('CLIENT NAME : ', '')
    list3.append(j)
  elif 'amount' in i.lower():
    j = i.replace('AMOUNT : ', '')
    list4.append(j)
  else:
    j = i.replace('TRANSACTION REF. : ', '')
    list5.append(j)
```

```python
data
```

```python
dict_h = dict()
dict_h['transaction date'] = list1  
dict_h['Client ID'] = list2
dict_h['Client Name'] = list3
dict_h['Amount'] = list4
dict_h['Transaction Ref.'] = list5
```

```python
import pandas as pd
df = pd.DataFrame(dict_h)
df
```
